# **Planteamiento del ejercicio 2: Aproximación de funciones de activación mediante series de Taylor**

```python
import numpy as np
import matplotlib.pyplot as plt

# Funciones reales
sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = np.tanh

# Series de Taylor
def taylor_sigmoid(x, n):
    return 0.5 + x/4 - (x**3)/48 * (n >= 3) + (x**5)/480 * (n >= 5)

def taylor_tanh(x, n):
    return x - (x**3)/3 * (n >= 2) + (2*x**5)/15 * (n >= 3) - (17*x**7)/315 * (n >= 4)

# Datos
x = np.linspace(-5, 5, 100)

# Cálculos
sig_real, tanh_real = sigmoid(x), tanh(x)
sig_t1, sig_t5 = taylor_sigmoid(x, 1), taylor_sigmoid(x, 5)
tanh_t1, tanh_t4 = taylor_tanh(x, 1), taylor_tanh(x, 4)

# Gráficos
plt.figure(figsize=(10, 4))

plt.subplot(121); plt.plot(x, sig_real, 'k-', x, sig_t1, 'b--', x, sig_t5, 'r--')
plt.title('Sigmoide')
plt.legend(['Real', '1 término', '5 términos'])
plt.grid()

plt.subplot(122)
plt.plot(x, tanh_real, 'k-', x, tanh_t1, 'b--', x, tanh_t4, 'r--')
plt.title('Tanh')
plt.legend(['Real', '1 término', '4 términos'])
plt.grid()

plt.tight_layout()
plt.show()

# Comparación en x = 1
x_p = 1.0
print(f"Sigmoide en x={x_p}: Real={sigmoid(x_p):.4f}, T1={taylor_sigmoid(x_p, 1):.4f}, T5={taylor_sigmoid(x_p, 5):.4f}")
print(f"Tanh en x={x_p}: Real={tanh(x_p):.4f}, T1={taylor_tanh(x_p, 1):.4f}, T4={taylor_tanh(x_p, 4):.4f}")
```
